## TCX Example

This example shows how easy it is to get heart rate zone info from downloaded TCX files from MapMyRide, Garmin and other fitness apps that support the TCX file format (actually Garmin's Fitness Database format).

```python
from ghettotcx import HeartRate, LatLong, TCX
```

```python
h = HeartRate("./example_data/example.tcx")
h.plot_heartrate()  # time series of heart rate readings
h.plot_heartrate_histogram() 
h.plot_heartratezone()  # heart rate zones are hard-coded. TODO: make user-defineable
```

Can also plot a panel of plots given a directory with many TCX files in it.  The ``create_heartrate_panel`` method will load in a list of HeartRate objects.  You can use the ``load_directory`` to quickly load these for you from a given directory. Make sure to specify ``HeartRate`` as the ``class_factory``.


```python
HeartRate.create_heartrate_panel(TCX.load_directory("./example_data", class_factory=HeartRate), fig_size=(6,2))
```

(Note: the examples_data directory only contains a single TCX file, so only one set of plots will be drawn).
